```python
def change_parameters(originality, brand, sex, category, year, month):

        df_update = df[df['originality'].isin(originality) & df['brand'].isin(brand) & df['sex'].isin(sex) & df['category'].isin(category) & df['year'].isin(year)& df['month'].isin(month)]

        for _, row in df_update.iterrows():
           fig =  px.scatter(df_update, x="pieces", y="retail_EUR", color="originality")
           fig2 = px.bar(df_update, x="brand", y="retail_EUR", barmode='stack',color='category', text = "pieces")
           fig3 = px.bar(df_update, x="category", y="retail_EUR", barmode='stack',color='brand', text = "pieces")
        fig.show()
        fig2.show()
        fig3.show()
```

```python
ui = widgets.HBox(
        [originality, 
         brand,
          sex, 
          category,
         year,
         month
        ], 
        layout=Layout(display='flex', flex_flow='row wrap', justify_content='space-between')
    )
```

```python
out = widgets.interactive_output(
    change_parameters, 
    {'originality': originality,
    'brand': brand, 
    'sex': sex, 
    'category': category,
    'year': year,
     'month': month,
    }
      )
```

```python
display(ui, out)
```
